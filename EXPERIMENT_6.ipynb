```python
#2 Check whether a number is even, odd, or zero using if statement

num = int(input("Enter a number: "))

if num == 0:
    print("The number is zero.")
elif num % 2 == 0:
    print("The number is even.")
else:
    print("The number is odd.")
```

Q6) Problem statement:

Write a python program to calculate the gross salary of an employee based on

If basic <= 10000: HRA = 20%, DA = 80%

If basic <= 20000: HRA = 25%, DA = 90%

OTHERWISE: HRA = 30%, DA = 95%

```python
basic_salary = float(input("Enter the basic salary: "))

if basic_salary <= 10000:
    hra_rate = 0.20
    da_rate = 0.80
elif basic_salary <= 20000:
    hra_rate = 0.25
    da_rate = 0.90
else:
    hra_rate = 0.30
    da_rate = 0.95

hra = basic_salary * hra_rate
da = basic_salary * da_rate
gross_salary = basic_salary + hra + da

print(f"Basic Salary: {basic_salary}")
print(f"HRA ({hra_rate*100}%): {hra}")
print(f"DA ({da_rate*100}%): {da}")
print(f"Gross Salary: {gross_salary}")
```

Q7) Write a python program to calculate income tax based on annual income:

*   upto $250,000: No tax

*   $250,001 - $500,000: 5%

*   $500,001 - $1,000,000: 20%

*   above $1,000,000: 30%

```python
annual_income = float(input("Enter your annual income: "))

if annual_income <= 250000:
    tax_rate = 0
    tax_amount = 0
elif annual_income <= 500000:
    tax_rate = 0.05
    tax_amount = (annual_income - 250000) * tax_rate
elif annual_income <= 1000000:
    tax_rate = 0.20

    tax_amount = (250000 * 0.05) + ((annual_income - 500000) * tax_rate)
else:
    tax_rate = 0.30

    tax_amount = (250000 * 0.05) + (500000 * 0.20) + ((annual_income - 1000000) * tax_rate)

print(f"Annual Income: ${annual_income:,.2f}")
print(f"Applicable Tax Rate: {tax_rate*100}%")
print(f"Calculated Income Tax: ${tax_amount:,.2f}")
```

**Q8) Write a python program for a given letter wheter it is a vowel or a consonant**

```python
letter = input("Enter a letter: ")

if len(letter) == 1 and letter.isalpha():
    if letter in ('a', 'e', 'i', 'o', 'u'):
        print(f"The letter '{letter}' is a vowel.")
    else:
        print(f"The letter '{letter}' is a consonant.")
else:
    print("Please enter a single alphabet character.")
```

**Q9) Write a python program to increment the date entered by user by 1 day.**

```python
date_str = input("Enter a date in YYYY-MM-DD format: ")

try:
    year_str, month_str, day_str = date_str.split('-')
    year = int(year_str)
    month = int(month_str)
    day = int(day_str)

    # Basic validation for month and day ranges (can be expanded)
    if not (1 <= month <= 12 and 1 <= day <= 31): # Simple check, full validation is complex with if-else
        print("Invalid date components (month or day out of typical range).")
    else:
        # Increment the day by 1
        day += 1

        # Determine the number of days in the current month
        days_in_current_month = 0
        if month in (1, 3, 5, 7, 8, 10, 12):
            days_in_current_month = 31
        elif month in (4, 6, 9, 11):
            days_in_current_month = 30
        elif month == 2: # February
            # Check for a leap year
            if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
                days_in_current_month = 29
            else:
                days_in_current_month = 28

        # Handle day rollover to the next month
        if day > days_in_current_month:
            day = 1 # Reset day to 1 for the new month
            month += 1 # Increment month

            # Handle month rollover to the next year
            if month > 12:
                month = 1 # Reset month to January
                year += 1 # Increment year

        print(f"The original date was: {year_str}-{month_str}-{day_str}")
        print(f"The date after incrementing by 1 day is: {year:04d}-{month:02d}-{day:02d}")

except ValueError:
    print("Invalid date format or components. Please ensure the date is YYYY-MM-DD and values are valid integers.")
except IndexError:
    print("Invalid date format. Please enter the date in YYYY-MM-DD format.")
```
